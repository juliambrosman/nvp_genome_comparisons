```python
%%R
col_factor=all$dbscan_grps
set_col_vec<-function(color_factor){
  custom_cols=c("#C044E8",
                "#70D04F",
                "#C74938",
                "#63B2F3",
                "#53703E",
                "#C35883",
                "#4D4169",
                "#8AD2AA",
                "#C68E53",
                "#C9C74E",
                "#8876CD",
                "#2D2DA7",
                "#58352F")
  num<-col_factor%>%as.factor%>%unique%>%length+100
  col_vec<-c("#95A192", rep(custom_cols, round(num/length(custom_cols))))
  return(col_vec)
}
```

```python
%%R -w 600 -h 500

col_vec<-set_col_vec(all$dbscan_groups)
db_scan_cols<-col_vec

ggplot(all, aes(tSNE1, tSNE2))+geom_point(aes(col=as.factor(dbscan_grps)), cex=3)->p
p<-p+scale_colour_manual(values=col_vec)+guides(col = guide_legend(nrow = 13))+theme_bw()+
  theme(line = element_blank(), axis.text = element_blank())+labs(title="tSNE Representation of Phage Genomes \nColored by DBScan Group", colour="DBScan Group", y=NULL, x=NULL)
plot(p)
```
